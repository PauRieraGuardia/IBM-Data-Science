To generate a table with information about the users, we can write a function containing all esirable parameters. For example, name, gender, city, etc. The parameters will depend on the requirements of the test to be performed. We call the Get Methods, listed at the beginning of this notebook. Then, we return pandas dataframe with the users.

```python
def get_users():
    users =[]
     
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
      
    return pd.DataFrame(users)     
```

### Example 2: Fruityvice API

Anothcer, more common woy to use APIs, is through <code>requests</code> library. The next lab, Requests and HTTP, will contain more information about requests.

We will start by importing all required libraries.

```python
import requests 
import json
```

We will ontain the frutyvice API data <code>requests-get("url")</code> function. The data is in a json format.

```python
data = requests.get("https://fruityvice.com/api/fruit/all")
```

We will retrieve results using <code>json.loads(/)</code> function

```python
results = json.loads(data.text)
```

We will convert our json data into pandas data frame.

```python
pd.DataFrame(results)
```

The result is in a nested json format. The nutrition column contains multiple subcolumbns, so the data to be "flattened" or normalized.

```python
df2 = pd.json_normalize(results)
```

```python
df2
```

Let's see if we can extract some informatiom from this dataframe. Perhaps, we need to know the family and genus of a cherry.

```python
cherry = df2.loc[df2["name"]=="Cherry"]
cherry
                      
```
