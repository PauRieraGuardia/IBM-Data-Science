### Practice Project: GDP Data extraction and processing

In this practice projecy we will extract data from a website using webscraping and request APIs process it using Pandas and numpy libraris.

We are tasked with creating a script that can extract the list of the top 10 largest economies of the world in descending order of their GDPs in Billion USD (rounded to 2 decimal places), as logged by the International Monetary Fund (IMF). 

The required data seems to be available on the URL mentioned below:


URL: https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29

#### Importing required Libraries 

```python
import numpy as np
import pandas as pd
```

### Exercises 

#### Exercise 1

Extract the required GDP data from the given URL using Web Scapring

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
### Extract tables from webpage using Panfas. Retain table number 3 as the required dataframe.

tables = pd.read_html(URL)
df= tables[3]
df.head()
```

```python
# Replace the column headers with column number

df.columns = range(df.shape[1])
df.head()
```

```python
# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
df = df[[0,2]]
df.head()
```

```python
# Retain the Rows woth index 1 to 10, indicating the top 10 economies o the world
df=df.iloc[1:11,:]
df.head()
```

```python
# Assign columns names as "Country" and "GDP (Million USD)"
df.columns = ["Country","GDP (Million USD)"]
df
```

### Exercise 2

Modify the GDP column of the DataFrame, converting the value available in Million USD to Billion USD. Use the <code>round()</code> method of Numpy libraryto round the value to 2 decimal places. Modify the header of the DataFrame to <code>GDP (Billion USD)</code>.

```python
# Change the data type of the "GDP "Million USD)" column to integer. Use astype() method.
df["GDP (Million USD)"] = df["GDP (Million USD)"].astype(int)
df.head()
```

```python
# Convert the GDP value in Million USD to Billion USD
df[["GDP (Million USD)"]] = df[["GDP (Million USD)"]]/1000
df.head()
```

```python
# use numpy.roud() method to round the value to 2 decimal places.

df[["GDP (Million USD)"]] = np.round(df[["GDP (Million USD)"]],2)
df.head()
```

```python
# Rename the column header from "GDP (Million USD)" to "GDP (Billion USD)"

df = df.rename(columns = {"GDP (Million USD)" : "GDP (Billion USD"})
df
```
